## Has the amount of precipitation in North Carolina increased since 1900?
	
    To explore this question, I decided to look at the average annual precipitation in three cities across the state of       North Carolina: Greenville, Chapel Hill and Asheville. I obtained climate data going back to 1900 from the National     Oceanic and Atmospheric Administration. After extracting the respective data for each city, I used it to create three   scatter plots. 

Visual inspection of the charts did not immediately reveal a pattern relevant to my question. I then calculated a line of best-fit for each plot and displayed it over the chart; in addition, I calculated the correlation coefficient for each citys data set. Greenville, Chapel Hill and Asheville have R-values of about 0.23, 0.10 and .07, respectively. These values, in concert with the apparent but very mild slopes to each linear regression line, suggest that the average annual precipitation has increased over time; however, the relationship is weak and we cannot make this statement with a great deal of confidence. Of additional note is the fact that the strength of the correlation seems to trend up as the datas origin nears the coast, suggesting a more emphatic increase in precipitation is, in fact, happening in coastal regions of the state.
