```python
#import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
import datetime
from pprint import pprint
import requests
```

```python
#import data

url = 'https://www.ncdc.noaa.gov/billions/events-NC-1980-2020.json'
data = requests.get(url).json()
pprint(data)
```

```python
#define the startdates and costs as lists 
#loop through the data, extracting dates and costs and appending them to the apropriate lists, converting them into integers
#fix the costs for readability

startdates=[]
costs=[]
for storm in data["data"]:
    Date=storm["begDate"]
    convertedDate=datetime.date(int(str(Date)[0:4]),int(str(Date)[4:6]),int(str(Date)[6:]))
    startdates.append(convertedDate)
    costs.append(float(storm["totalCost"])/1000)
    
    
    
```

```python
#print costs
costs
```

```python
#sort & print the dates 
startdates=sorted(startdates)
```

```python
#plot the sum of damage done greater than a billion dollars for each year
#label the graph
#save as png
plt.plot(startdates,costs)
plt.xlabel('Year')
plt.ylabel('Damage in Billions of Dollars')
plt.title("North Carolina's Billion Dollar Storms")
plt.savefig("billion_graph.png")
```
