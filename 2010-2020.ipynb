```python
# Group data by year

print(wind_df["Observation Time"].unique())

wind_groupby_year = wind_df.groupby("Observation Time")
```

```python
# Rename columns as "YEAR" and "MONTH"

year_and_month_df = year_and_month_df.rename(columns= {0: "YEAR", 1: "MONTH"})
```
